# Task
Konversi kolom 'hps', 'nilai_penawaran', 'nilai_terkoreksi', 'nilai_negosiasi', dan 'nilai_kontrak' menjadi tipe data numerik (float), dengan menangani koma sebagai pemisah desimal, lalu tampilkan informasi DataFrame.

## Konversi Tipe Data Numerik

### Subtask:
Konversi kolom-kolom seperti 'hps', 'nilai_penawaran', 'nilai_terkoreksi', 'nilai_negosiasi', dan 'nilai_kontrak' dari tipe objek menjadi tipe numerik (float). Perhatikan bahwa beberapa nilai mungkin menggunakan koma sebagai pemisah desimal, jadi pastikan untuk menanganinya dengan benar.


**Reasoning**:
The user wants to convert specific columns to numeric type, handling comma decimal separators. I need to define the columns, iterate through them, replace commas with dots, and then convert to float, coercing errors.



```python
#@title Step 1. Konversi Kolom Finansial ke Numerik
print("=== KONVERSI KOLM FINANSIAL ===")

columns_to_convert = ['hps', 'nilai_penawaran', 'nilai_terkoreksi', 'nilai_negosiasi', 'nilai_kontrak']

for col in columns_to_convert:
    # Ubah kolom menjadi string agar fungsi .str.replace() dapat digunakan
    df[col] = df[col].astype(str)
    # Ganti tanda koma menjadi titik untuk format desimal
    df[col] = df[col].str.replace(',', '.')
    # Ubah nilai menjadi numerik; jika gagal akan diubah menjadi NaN
    df[col] = pd.to_numeric(df[col], errors='coerce')

print("=== UPDATED DATAFRAME INFO ===")
df.info()
```

## Feature Engineering (Mengubah data mentah menjadi fitur yang lebih relevan dan bermakna) : Rasio Penawaran terhadap HPS

### Subtask:
Buat fitur baru yang relevan dengan tujuan proyek: 'persentase_penurunan_hps'. Fitur ini akan menghitung persentase penurunan harga penawaran relatif terhadap HPS. Rumusnya adalah ((HPS - nilai_penawaran) / HPS) * 100.


**Reasoning**:
Untuk menghitung persentase penurunan HPS, pertama-tama nilai HPS yang nol diganti dengan kosong (NaN) supaya perhitungan nggak error. Kalau nanti muncul nilai tak hingga (inf), juga diganti dengan kosong supaya datanya rapi. Terakhir, lihat lima baris pertama untuk memastikan kolom baru sudah benar.



```python
#@title Step 1. Menghitung Persentase Penurunan HPS
print("=== DATAFRAME DENGAN KOLOM 'PERSENTASE_PENURUNAN_HPS' ===")
# Ganti nilai 0 pada HPS dengan NaN untuk menghindari pembagian dengan nol
df['hps_cleaned'] = df['hps'].replace(0, np.nan)
# Hitung persentase penurunan HPS dibandingkan dengan nilai penawaran
df['persentase_penurunan_hps'] = ((df['hps_cleaned'] - df['nilai_penawaran']) / df['hps_cleaned']) * 100
# Tangani nilai tak hingga (infinite) akibat pembagian
df['persentase_penurunan_hps'] = df['persentase_penurunan_hps'].replace([np.inf, -np.inf], np.nan)
# Tampilkan beberapa baris untuk kolom baru
print(df[['hps', 'nilai_penawaran', 'persentase_penurunan_hps']].head())
```

```python
%%writefile app.py

import streamlit as st
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
from datetime import datetime
import numpy as np

# Set page configuration
st.set_page_config(
    page_title="Dashboard Tender Usaha Kecil 2023-2024",
    page_icon="",
    layout="wide",
    initial_sidebar_state="expanded"
)

# Load data
@st.cache_data
def load_data():
    # Load the CSV data
    df = pd.read_csv('cleaned_datatender.csv')

    # Clean numeric columns - handle comma as decimal separator
    numeric_columns = ['pagu', 'hps', 'nilai_penawaran', 'nilai_terkoreksi',
                     'nilai_negosiasi', 'nilai_kontrak']

    for col in numeric_columns:
        if col in df.columns:
            df[col] = df[col].astype(str).str.replace(',', '.', regex=False)
            df[col] = pd.to_numeric(df[col], errors='coerce')

    # Convert tahun_anggaran to integer
    df['tahun_anggaran'] = df['tahun_anggaran'].astype(str)

    return df

def main():
    st.title("Dashboard Monitoring Tender Kualifikasi Usaha Kecil 2023-2024")
    st.markdown("---")

    # Load data
    try:
        df = load_data()
    except Exception as e:
        st.error(f"Error loading data: {e}")
        return

    # Sidebar filters
    st.sidebar.header("Filter Data")

    # Tahun Anggaran filter
    tahun_options = ['All'] + sorted(df['tahun_anggaran'].unique().tolist())
    selected_tahun = st.sidebar.selectbox("Tahun Anggaran", tahun_options)

    # Jenis KLPD filter
    jenis_klpd_options = ['All'] + sorted(df['jenis_klpd'].unique().tolist())
    selected_jenis_klpd = st.sidebar.selectbox("Jenis KLPD", jenis_klpd_options)

    # Sumber Dana filter
    sumber_dana_options = ['All'] + sorted(df['sumber_dana'].dropna().unique().tolist())
    selected_sumber_dana = st.sidebar.selectbox("Sumber Dana", sumber_dana_options)

    # Status Tender filter
    status_options = ['All'] + sorted(df['status_tender'].dropna().unique().tolist())
    selected_status = st.sidebar.selectbox("Status Tender", status_options)

    # Apply filters
    filtered_df = df.copy()
    if selected_tahun != 'All':
        filtered_df = filtered_df[filtered_df['tahun_anggaran'] == selected_tahun]
    if selected_jenis_klpd != 'All':
        filtered_df = filtered_df[filtered_df['jenis_klpd'] == selected_jenis_klpd]
    if selected_sumber_dana != 'All':
        filtered_df = filtered_df[filtered_df['sumber_dana'] == selected_sumber_dana]
    if selected_status != 'All':
        filtered_df = filtered_df[filtered_df['status_tender'] == selected_status]

    # Main dashboard
    col1, col2, col3, col4 = st.columns(4)

    with col1:
        total_tender = len(filtered_df)
        st.metric("Total Paket Tender", f"{total_tender:,}")

    with col2:
        total_pagu = filtered_df['pagu'].sum() / 1e9  # Convert to billions
        st.metric("Total Pagu (Rp)", f"Rp {total_pagu:,.2f} M")

    with col3:
        total_kontrak = filtered_df['nilai_kontrak'].sum() / 1e9  # Convert to billions
        st.metric("Total Nilai Kontrak (Rp)", f"Rp {total_kontrak:,.2f} M")

    with col4:
        avg_efisiensi = ((filtered_df['pagu'] - filtered_df['nilai_kontrak']) / filtered_df['pagu'] * 100).mean()
        st.metric("Rata-rata Efisiensi", f"{avg_efisiensi:.1f}%")

    st.markdown("---")

    # Tabs for different analyses
    tab1, tab2, tab3, tab4, tab5, tab6 = st.tabs([
        "Overview",
        "Berdasarkan Instansi",
        "Analisis Nilai",
        "Metode & Status",
        "Analisis Penyedia",
        "Data Detail"
    ])

    with tab1:
        st.subheader("Overview Tender Usaha Kecil")

        # Row 1: High-Level KPIs
        col1, col2, col3 = st.columns(3)

        with col1:
            # Total Tender per Tahun (Line Chart)
            tender_per_tahun = filtered_df['tahun_anggaran'].value_counts().sort_index()
            fig_tender_tahun = px.line(
                x=tender_per_tahun.index,
                y=tender_per_tahun.values,
                title="Total Tender per Tahun",
                labels={'x': 'Tahun Anggaran', 'y': 'Jumlah Tender'},
                markers=True
            )
            st.plotly_chart(fig_tender_tahun, use_container_width=True)

        with col2:
            # Persentase Tender Selesai vs Gagal (Donut Chart)
            if 'status_tender' in filtered_df.columns:
                status_count = filtered_df['status_tender'].value_counts()
                fig_status = px.pie(
                    values=status_count.values,
                    names=status_count.index,
                    title="Persentase Status Tender",
                    hole=0.4
                )
                st.plotly_chart(fig_status, use_container_width=True)

        with col3:
            # Distribution by Jenis KLPD
            jenis_klpd_count = filtered_df['jenis_klpd'].value_counts()
            fig_klpd = px.pie(
                values=jenis_klpd_count.values,
                names=jenis_klpd_count.index,
                title="Distribusi Berdasarkan Jenis KLPD"
            )
            st.plotly_chart(fig_klpd, use_container_width=True)

        # Row 2
        col1, col2 = st.columns(2)

        with col1:
            # Distribution by Sumber Dana
            sumber_dana_count = filtered_df['sumber_dana'].value_counts().head(10)
            fig_dana = px.bar(
                x=sumber_dana_count.values,
                y=sumber_dana_count.index,
                orientation='h',
                title="Top 10 Sumber Dana",
                labels={'x': 'Jumlah Tender', 'y': 'Sumber Dana'}
            )
            st.plotly_chart(fig_dana, use_container_width=True)

        with col2:
            # Distribution of Jenis Pengadaan
            if 'jenis_pengadaan' in filtered_df.columns:
                jenis_pengadaan_count = filtered_df['jenis_pengadaan'].value_counts()
                fig_jenis = px.pie(
                    values=jenis_pengadaan_count.values,
                    names=jenis_pengadaan_count.index,
                    title="Distribusi Jenis Pengadaan",
                    hole=0.4
                )
                st.plotly_chart(fig_jenis, use_container_width=True)

    with tab2:
        st.subheader("Analisis Berdasarkan Instansi")

        col1, col2 = st.columns(2)

        with col1:
            # Top KLPD by jumlah tender
            top_klpd = filtered_df['nama_klpd'].value_counts().head(10)
            fig_top_klpd = px.bar(
                x=top_klpd.values,
                y=top_klpd.index,
                orientation='h',
                title="Top 10 KLPD by Jumlah Tender",
                labels={'x': 'Jumlah Tender', 'y': 'KLPD'}
            )
            st.plotly_chart(fig_top_klpd, use_container_width=True)

        with col2:
            # Top Satker by nilai kontrak
            top_satker = filtered_df.groupby('nama_satker')['nilai_kontrak'].sum().sort_values(ascending=False).head(10)
            fig_top_satker = px.bar(
                x=top_satker.values / 1e9,  # Convert to billions
                y=top_satker.index,
                orientation='h',
                title="Top 10 Satker by Nilai Kontrak (Rp Miliar)",
                labels={'x': 'Nilai Kontrak (Rp Miliar)', 'y': 'Satuan Kerja'}
            )
            st.plotly_chart(fig_top_satker, use_container_width=True)

        # Nilai Kontrak Berdasarkan Metode Pemilihan
        st.subheader("Nilai Kontrak Berdasarkan Metode Pemilihan")
        if 'mtd_pemilihan' in filtered_df.columns:
            metode_nilai = filtered_df.groupby('mtd_pemilihan')['nilai_kontrak'].sum().sort_values(ascending=False)
            fig_metode_nilai = px.bar(
                x=metode_nilai.index,
                y=metode_nilai.values / 1e9,
                title="Total Nilai Kontrak per Metode Pemilihan (Rp Miliar)",
                labels={'x': 'Metode Pemilihan', 'y': 'Nilai Kontrak (Rp Miliar)'}
            )
            st.plotly_chart(fig_metode_nilai, use_container_width=True)

    with tab3:
        st.subheader("Analisis Nilai Tender")

        col1, col2 = st.columns(2)

        with col1:
            # Distribution of pagu values
            fig_pagu = px.histogram(
                filtered_df,
                x='pagu',
                nbins=50,
                title="Distribusi Nilai Pagu",
                labels={'pagu': 'Nilai Pagu (Rp)'}
            )
            fig_pagu.update_xaxes(tickformat=".2s")
            st.plotly_chart(fig_pagu, use_container_width=True)

        with col2:
            # Scatter plot: Pagu vs Nilai Kontrak
            scatter_data = filtered_df.dropna(subset=['pagu', 'nilai_kontrak'])
            fig_scatter = px.scatter(
                scatter_data,
                x='pagu',
                y='nilai_kontrak',
                title="Pagu vs Nilai Kontrak",
                labels={'pagu': 'Pagu (Rp)', 'nilai_kontrak': 'Nilai Kontrak (Rp)'},
                trendline="ols"
            )
            fig_scatter.update_xaxes(tickformat=".2s")
            fig_scatter.update_yaxes(tickformat=".2s")
            st.plotly_chart(fig_scatter, use_container_width=True)

        # Waterfall Chart for Pagu-HPS-Penawaran-Kontrak
        st.subheader("Alur Nilai: Pagu  HPS  Penawaran  Kontrak")
        if all(col in filtered_df.columns for col in ['pagu', 'hps', 'nilai_penawaran', 'nilai_kontrak']):
            # Sample data for one tender or aggregated values
            sample_data = filtered_df[['pagu', 'hps', 'nilai_penawaran', 'nilai_kontrak']].mean()

            fig_waterfall = go.Figure(go.Waterfall(
                name="Nilai", orientation="v",
                measure=["absolute", "relative", "relative", "total"],
                x=["Pagu", "HPS", "Penawaran", "Kontrak"],
                textposition="outside",
                y=[sample_data['pagu'],
                   sample_data['hps'] - sample_data['pagu'],
                   sample_data['nilai_penawaran'] - sample_data['hps'],
                   sample_data['nilai_kontrak'] - sample_data['nilai_penawaran']],
                connector={"line":{"color":"rgb(63, 63, 63)"}},
            ))

            fig_waterfall.update_layout(
                title="Waterfall Chart: Alur Penurunan Nilai",
                showlegend=True
            )
            st.plotly_chart(fig_waterfall, use_container_width=True)

        # Efisiensi analysis
        st.subheader("Analisis Efisiensi")
        filtered_df_eff = filtered_df.dropna(subset=['pagu', 'nilai_kontrak'])
        filtered_df_eff = filtered_df_eff[filtered_df_eff['pagu'] > 0]
        filtered_df_eff['efisiensi'] = ((filtered_df_eff['pagu'] - filtered_df_eff['nilai_kontrak']) / filtered_df_eff['pagu']) * 100

        col1, col2 = st.columns(2)

        with col1:
            fig_efisiensi = px.histogram(
                filtered_df_eff,
                x='efisiensi',
                nbins=30,
                title="Distribusi Tingkat Efisiensi",
                labels={'efisiensi': 'Efisiensi (%)'}
            )
            st.plotly_chart(fig_efisiensi, use_container_width=True)

        with col2:
            # Efisiensi per Tahun
            if 'tahun_anggaran' in filtered_df_eff.columns:
                efisiensi_tahun = filtered_df_eff.groupby('tahun_anggaran')['efisiensi'].mean().reset_index()
                fig_efisiensi_tahun = px.line(
                    efisiensi_tahun,
                    x='tahun_anggaran',
                    y='efisiensi',
                    title="Rata-rata Efisiensi per Tahun",
                    labels={'tahun_anggaran': 'Tahun', 'efisiensi': 'Efisiensi (%)'},
                    markers=True
                )
                st.plotly_chart(fig_efisiensi_tahun, use_container_width=True)

    with tab4:
        st.subheader("Analisis Metode Pemilihan dan Status")

        col1, col2 = st.columns(2)

        with col1:
            # Distribution by metode pemilihan
            mtd_pemilihan_count = filtered_df['mtd_pemilihan'].value_counts()
            fig_mtd = px.pie(
                values=mtd_pemilihan_count.values,
                names=mtd_pemilihan_count.index,
                title="Distribusi Metode Pemilihan"
            )
            st.plotly_chart(fig_mtd, use_container_width=True)

        with col2:
            # Distribution by status tender
            status_count = filtered_df['status_tender'].value_counts()
            fig_status = px.bar(
                x=status_count.values,
                y=status_count.index,
                orientation='h',
                title="Distribusi Status Tender",
                labels={'x': 'Jumlah', 'y': 'Status'}
            )
            st.plotly_chart(fig_status, use_container_width=True)

        # Success rate by metode pemilihan
        st.subheader("Tingkat Keberhasilan Berdasarkan Metode")
        success_by_method = filtered_df.groupby('mtd_pemilihan')['status_tender'].apply(
            lambda x: (x == 'Selesai').sum() / len(x) * 100
        ).sort_values(ascending=False)

        fig_success = px.bar(
            x=success_by_method.values,
            y=success_by_method.index,
            orientation='h',
            title="Tingkat Keberhasilan Berdasarkan Metode Pemilihan (%)",
            labels={'x': 'Tingkat Keberhasilan (%)', 'y': 'Metode Pemilihan'}
        )
        st.plotly_chart(fig_success, use_container_width=True)

        # Analisis Koreksi Penawaran
        st.subheader("Analisis Koreksi Penawaran")
        if 'nilai_terkoreksi' in filtered_df.columns:
            koreksi_data = filtered_df[filtered_df['nilai_terkoreksi'].notna()]
            if len(koreksi_data) > 0:
                fig_koreksi = px.histogram(
                    koreksi_data,
                    x='nilai_terkoreksi',
                    title="Distribusi Nilai Terkoreksi",
                    labels={'nilai_terkoreksi': 'Nilai Terkoreksi (Rp)'}
                )
                fig_koreksi.update_xaxes(tickformat=".2s")
                st.plotly_chart(fig_koreksi, use_container_width=True)

    with tab5:
        st.subheader("Analisis Penyedia")

        col1, col2 = st.columns(2)

        with col1:
            # Top Penyedia berdasarkan Nilai Kontrak
            if 'kd_penyedia' in filtered_df.columns:
                top_penyedia = filtered_df.groupby('kd_penyedia')['nilai_kontrak'].sum().sort_values(ascending=False).head(10)
                fig_top_penyedia = px.bar(
                    x=top_penyedia.values / 1e9,
                    y=top_penyedia.index,
                    orientation='h',
                    title="Top 10 Penyedia berdasarkan Nilai Kontrak (Rp Miliar)",
                    labels={'x': 'Nilai Kontrak (Rp Miliar)', 'y': 'Kode Penyedia'}
                )
                st.plotly_chart(fig_top_penyedia, use_container_width=True)

        with col2:
            # Perbandingan Nilai Penawaran vs Nilai Kontrak
            if all(col in filtered_df.columns for col in ['nilai_penawaran', 'nilai_kontrak']):
                scatter_penawaran = filtered_df.dropna(subset=['nilai_penawaran', 'nilai_kontrak'])
                fig_penawaran_kontrak = px.scatter(
                    scatter_penawaran,
                    x='nilai_penawaran',
                    y='nilai_kontrak',
                    title="Perbandingan Nilai Penawaran vs Nilai Kontrak",
                    labels={'nilai_penawaran': 'Nilai Penawaran (Rp)', 'nilai_kontrak': 'Nilai Kontrak (Rp)'},
                    trendline="ols"
                )
                fig_penawaran_kontrak.update_xaxes(tickformat=".2s")
                fig_penawaran_kontrak.update_yaxes(tickformat=".2s")
                st.plotly_chart(fig_penawaran_kontrak, use_container_width=True)

        # Boxplot for efisiensi by penyedia
        st.subheader("Distribusi Efisiensi per Penyedia (Top 20)")
        if 'kd_penyedia' in filtered_df_eff.columns:
            top_penyedia_list = filtered_df_eff.groupby('kd_penyedia')['nilai_kontrak'].sum().nlargest(20).index
            filtered_top_penyedia = filtered_df_eff[filtered_df_eff['kd_penyedia'].isin(top_penyedia_list)]

            fig_box_penyedia = px.box(
                filtered_top_penyedia,
                x='kd_penyedia',
                y='efisiensi',
                title="Distribusi Efisiensi untuk Top 20 Penyedia",
                labels={'kd_penyedia': 'Kode Penyedia', 'efisiensi': 'Efisiensi (%)'}
            )
            st.plotly_chart(fig_box_penyedia, use_container_width=True)

    with tab6:
        st.subheader("Data Detail Tender")

        # Show raw data with filters
        col1, col2 = st.columns(2)

        with col1:
            show_columns = st.multiselect(
                "Pilih kolom yang ditampilkan:",
                options=filtered_df.columns.tolist(),
                default=['nama_paket', 'nama_klpd', 'nama_satker', 'pagu', 'nilai_kontrak', 'status_tender']
            )

        with col2:
            rows_to_show = st.slider("Jumlah baris yang ditampilkan:", 10, 100, 20)

        if show_columns:
            st.dataframe(
                filtered_df[show_columns].head(rows_to_show),
                use_container_width=True
            )

        # Download filtered data
        csv = filtered_df.to_csv(index=False)
        st.download_button(
            label="Download Data Filtered (CSV)",
            data=csv,
            file_name=f"tender_usaha_kecil_filtered_{datetime.now().strftime('%Y%m%d')}.csv",
            mime="text/csv"
        )

    # Footer
    st.markdown("---")
    st.markdown(
        "**Dashboard Tender Kualifikasi Usaha Kecil** | "
        "Data periode 2023-2024 | "
        "Dibuat dengan Streamlit"
    )

if __name__ == "__main__":
    main()
```
